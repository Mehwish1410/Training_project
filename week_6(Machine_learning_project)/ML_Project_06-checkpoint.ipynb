```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
#Handle outliers
Q1= df['Height'].quantile(0.25)
Q3= df['Height'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q1 + 1.5*IQR
df = df[(df["Height"]>lower) & (df["Height"]<upper)]

#Handle outliers
Q1= df['Weight'].quantile(0.25)
Q3= df['Weight'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q1 + 1.5*IQR
df = df[(df["Weight"]>lower) & (df["Weight"]<upper)]
df.shape
```

o	Encode categorical features and the target variable.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])  
df['family_history_with_overweight'] = le.fit_transform(df['family_history_with_overweight'])  
df['FAVC'] = le.fit_transform(df['FAVC'])  
df['SMOKE'] = le.fit_transform(df['SMOKE'])  
df['CAEC'] = le.fit_transform(df['CAEC'])  
df['CALC'] = le.fit_transform(df['CALC'])  
df['MTRANS'] = le.fit_transform(df['MTRANS'])  
df['NObeyesdad'] = le.fit_transform(df['NObeyesdad'])  
df
```

# 2.	Feature Selection using PCA:

o	Apply PCA to reduce the dimensionality of the dataset.

o Select the top 6 principal components that explain the majority of variance.

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Separate features and target variable
X = df.drop(columns=['NObeyesdad'])
y = df['NObeyesdad']

# Apply PCA
pca = PCA(n_components=6)
X_pca = pca.fit_transform(X)

loadings = pca.components_.T  

feature_namess = X.columns
loadings_df = pd.DataFrame(loadings, columns=['PC1', 'PC2','PC3','PC4','PC5','PC6'], index=feature_namess)
```

```python
influential_features = {}
for pc in loadings_df.columns:
    influential_features[pc] = loadings_df[pc].abs().idxmax()
```

```python
print("Most influential original features for each principal component:")
for pc, feature in influential_features.items():
    print(f"{pc}: {feature}")
```
