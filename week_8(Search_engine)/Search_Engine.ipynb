```python
# Function to search for a product based on name and brand
def find_product_description(product_name, brand_name):
    # Search for the product in the DataFrame
    matching_product = df[(df['Product_Name'].str.contains(product_name, case=False, na=False)) & 
                          (df['Brand'].str.contains(brand_name, case=False, na=False))]
    
    if not matching_product.empty:
        # Return description if a match is found
        return matching_product[['Product_Name', 'Brand', 'Category','Price','Ratings','Reviews Count','Availability']]
    else:
        return 'Product not found'
        
```

```python
# Example usage with user input
product_name = input("Enter the product name: ")
brand_name = input("Enter the brand name: ")
```

```python
# Get the product description based on user input
product_description = find_product_description(product_name, brand_name)
```
