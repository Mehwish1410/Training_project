```python
from sklearn.feature_selection import chi2,SelectKBest
import pandas as pd
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#binning
X['EDUCATED']=pd.cut(X['EDUCATED'], bins=3, labels=['small','medium','large'])
X['NEVER EDUCATED']=pd.cut(X['NEVER EDUCATED'], bins=3, labels=['small','medium','large'])
X['JOBS']=pd.cut(X['JOBS'], bins=3, labels=['small','medium','large'])
X['TOTAL POPULATION']=pd.cut(X['TOTAL POPULATION'], bins=3, labels=['small','medium','large'])
X['AGE']=pd.cut(X['AGE'], bins=3, labels=['small','medium','large'])
X['ID']=pd.cut(X['ID'], bins=3, labels=['small','medium','large'])
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['TOTAL POPULATION']=le.fit_transform(X['TOTAL POPULATION'])
X['NEVER EDUCATED']=le.fit_transform(X['NEVER EDUCATED'])
X['EDUCATED']=le.fit_transform(X['EDUCATED'])
X['JOBS']=le.fit_transform(X['JOBS'])
X['ID']=le.fit_transform(X['ID'])
X['AGE']=le.fit_transform(X['AGE'])
```

```python
X
```

```python
from sklearn.feature_selection import chi2,SelectKBest
select= SelectKBest(chi2,k=3)
selected_data = select.fit_transform(X,y)
s_features = X.columns[select.get_support()].tolist()
s_features
slected_df = pd.DataFrame(selected_data,columns=s_features)
```
